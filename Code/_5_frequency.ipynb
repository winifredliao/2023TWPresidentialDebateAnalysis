```python
df_all=pd.read_csv('C:/Users/blues/Desktop/Projects/2024TWPresidentialDebateAnalysis/All_CSV/final_dataset.csv')
```

```python
import pandas as pd
from ckiptagger import WS
from collections import Counter

# 加载中文分词器
ws = WS("./data")
```

```python
word_to_weight = {
    "是": 1,
    "不是": 2,
    "辯論會": 2,
    "保柯": 5,
    "棄侯": 5,
    "賴皮寮": 1,
}
dictionary = construct_dictionary(word_to_weight)
print(dictionary)
```

```python
# 逐行读取并进行中文分词
tokenized_text_list = []
for index, row in df_all.iterrows():
    text=[str(row['text'])]
    word_sentence_list = ws(text,
    coerce_dictionary = dictionary)
    for words in word_sentence_list:
        for i in words:
            tokenized_text_list.append(i)
#print(tokenized_text_list)
```

```python
# 统计词频
word_freq_list = Counter(tokenized_text_list)
#print(word_freq_list)
```

```python
df_freq = pd.DataFrame(list(word_freq_list.items()), columns=['Element', 'Frequency']).sort_values(by='Frequency', ascending=False)
```
