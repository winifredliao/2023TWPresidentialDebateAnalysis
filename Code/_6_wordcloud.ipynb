```python
df=pd.read_csv('C:/Users/blues/Desktop/Projects/2024TWPresidentialDebateAnalysis/All_CSV/frequency.csv')
```

```python
data_dict = dict(zip(df['Element'], df['Frequency']))
data_dict
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

TC_FONT_PATH='C:/Users/blues/Desktop/Projects/2024TWPresidentialDebateAnalysis/Noto_Serif_TC/NotoSerifTC-Regular.otf'

word_freq = data_dict

wordcloud = WordCloud(width=1200, height=600, background_color='white', font_path=TC_FONT_PATH,).generate_from_frequencies(word_freq)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig('C:/Users/blues/Desktop/Projects/2024TWPresidentialDebateAnalysis/PNG/wordcloud.png')
```
